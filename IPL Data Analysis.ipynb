 ### 3.3 Preprocessing

- Dealing with missing values<br/>
    - Dropping the column __Umpire3__ as it has too many _null_ values.
    - Replacing __null__ values for __umpire1__ and __umpire2__ column.
    - Raplacing __null__ values for the __city__ column.
- Correctnes of data<br/>
    - Correcting Team name for __Pune's__ team.                                                                             
- New column Addition
    - Four new columns __Batting team__, __Fielding team__, __batting winner or Bowling Winner__, __toss_winner_is_winner__ added

```python
#ipl_data.umpire3
#ipl_data.umpire3.head() # as umpire3 have mostly values null so we will drop this column
ipl_data.drop('umpire3', axis = 1,inplace = True) # dropping umpire3 column
```

```python
# Analysing missing umpire1 and umpire2
ipl_data[pd.isnull(ipl_data['umpire1'])]
ipl_data.query('season==2017 and team1=="Royal Challengers Bangalore" and team2=="Delhi Daredevils"')['umpire1'].mode()[0]
ipl_data.query('season==2017 and team1=="Royal Challengers Bangalore" and team2=="Delhi Daredevils"')['umpire2'].mode()[0]

# as per above analysis we can replace the missing value of umpire1 within group 'season','team1','team2'
ipl_data_new=ipl_data.groupby(['season','team1','team2'])['umpire1'].transform(lambda x: x.value_counts(dropna=False).index[0]).to_frame()
ipl_data.loc[ipl_data['umpire1'].isnull().to_list(),'umpire1']=ipl_data_new[ipl_data['umpire1'].isnull().to_list()]

# as per above analysis we can replace the missing value of umpire2 within group 'season','team1','team2'
ipl_data_new1=ipl_data.groupby(['season','team1','team2'])['umpire2'].transform(lambda x: x.value_counts(dropna=False).index[0]).to_frame()
ipl_data.loc[ipl_data['umpire2'].isnull().to_list(),'umpire2']=ipl_data_new1[ipl_data['umpire2'].isnull().to_list()]
```

```python
# Analysing missing player of the match and winner
#ipl_data.loc[ipl_data['player_of_match'].isnull().to_list(),]
ipl_data.loc[ipl_data['winner'].isnull().to_list(),]
```

<br>By Analysing above data we come to conclusion that __winner__ and __player of the match__ is null only in those cases where match could not be completed due to any reason and ended up with no result i.e. bad weather.<br>

```python
# analysing null values for city
#ipl_data.loc[ipl_data['city'].isnull().to_list(),]# seems like all null values are for a particular season
#ipl_data.loc[ipl_data['venue']=='Dubai International Cricket Stadium'] .count()
#ipl_data.query('season==2014 & city!=city').count()
# all missing values for city are for seson 2014 so lets drill down further
ipl_data.query('season==2014')['city'].mode()[0]
```

As from the above analysis we can clearly say most of the matches are played out of india in __Abu Dhabi__ for season 2014.
so lets replace all null values of city column with __'Abu Dhabi'__<br><br>

```python
# Replacing null value with the mode of city column for season 2014
ipl_data['city'].fillna(ipl_data.query('season==2014')['city'].mode()[0], inplace = True)
#ipl_data.query('season==2014').head()
```

```python
# correct the team name and make the data homogenious for column team1,team2 and winner
ipl_data['team1']=ipl_data['team1'].replace('Supergiants', 'Supergiant', regex=True)
ipl_data['team2']=ipl_data['team2'].replace('Supergiants', 'Supergiant', regex=True)
ipl_data['winner']=ipl_data['winner'].replace('Supergiants', 'Supergiant', regex=True)
```

```python
# For tied matches set winner to null
ipl_data.loc[ipl_data['result'] =='tie', 'winner']= np.NaN
```

```python
# introducing four new columns Batting Team,Fielding Team ,batting winner or Bowling Winner,toss_winner_is_winner
ipl_data["Batting Team"]=ipl_data.apply(lambda row: row['toss_winner'] if row['toss_decision'] == 'bat' else (row['team2'] if row['toss_winner']==row['team1'] else row['team1']), axis=1)
ipl_data["Fielding Team"]=ipl_data.apply(lambda row: row['toss_winner'] if row['toss_decision'] == 'field' else (row['team2'] if row['toss_winner']==row['team1'] else row['team1']), axis=1)
ipl_data["batting winner or Bowling Winner"]=ipl_data.apply(lambda row: 'bat' if row['winner']==row['Batting Team'] else ('field' if row['winner']==row['Fielding Team'] else 'No result or tie'),axis=1)
ipl_data['toss_winner_is_winner'] = 'no'
ipl_data['toss_winner_is_winner'].loc[ipl_data.toss_winner == ipl_data.winner] = 'yes'
```

####  4.2.2 Win percentage at each venue by first batting or bowling? 

```python
df=ipl_data[['venue','city','winner','batting winner or Bowling Winner']]
df=df[df.winner.notnull()]
df=df[['venue','city','batting winner or Bowling Winner']]
df.columns=['Venue','City','BatOrField']
venue_total_match=df.groupby(["Venue","City"])[["City"]].count().rename(columns={"City":"MatchesPlayed"}).reset_index()
won_by_BatOrField=df.groupby(["Venue","City","BatOrField"])[["BatOrField"]].count().rename(columns={"BatOrField":"MatchesWon"}).reset_index()
df=pd.merge(venue_total_match,won_by_BatOrField,how='left',left_on=['Venue','City'],right_on=['Venue','City'])
percent_win=((df['MatchesWon']/df['MatchesPlayed'])*100).round(2).to_list()
df["PercentWin"]=percent_win
df.fillna(0,inplace = True)
df=df.query('PercentWin >= 70')
df.reset_index(inplace=True)
df[['Venue','City','BatOrField','MatchesPlayed','PercentWin']]
```

we come to know there are three grounds __Green Park Kanpur__,__Holkar Cricket Stadium Indore__ and __JSCA International Stadium Complex Ranchi__  where if you field first your win chances are greater than 70%
Let's plot the data for above three venues


```python
df=ipl_data.loc[ipl_data["venue"].isin(['Green Park','Holkar Cricket Stadium','JSCA International Stadium Complex'])]
f,ax = plt.subplots(1,3,figsize=(22,6))
df['batting_winner_or_Bowling_Winner'][df['venue'] == 'Green Park'].value_counts().plot.pie(explode=[0.1],colors=('#99ff99','#ffcc99'),autopct='%1.1f%%',ax=ax[0],shadow=True)
df['batting_winner_or_Bowling_Winner'][df['venue'] == 'Holkar Cricket Stadium'].value_counts().plot.pie(explode=[0.2,0],colors=('#99ff99','#6572c9'),autopct='%1.1f%%',ax=ax[1],shadow=True)
df['batting_winner_or_Bowling_Winner'][df['venue'] == 'JSCA International Stadium Complex'].value_counts().plot.pie(explode=[0.1,0],colors=('#99ff99','#6572c9'),autopct='%1.1f%%',ax=ax[2],shadow=True)
ax[0].set_title('Green Park')
ax[1].set_title('Holkar Cricket Stadium')
ax[2].set_title('JSCA International Stadium Complex')


```

#### 4.2.3 Win percentage of team venue wise?

```python
df1=ipl_data[['venue','city','team1','winner']]
df2=ipl_data[['venue','city','team2','winner']]
df2=df2[df2.winner.notnull()]
df1=df1[df1.winner.notnull()]
df2.columns = df1.columns
df=pd.concat([df1, df2],axis=0,sort=False,ignore_index=True)
df.columns=['Venue','City','Team','Winner']
team_total_match=df.groupby(["Venue","City","Team"])[["Team"]].count().rename(columns={"Team":"MatchesPlayed"}).reset_index()
team_total_match_won=ipl_data.groupby(["venue","city","winner"])[["winner"]].count().rename(columns={"winner":"MatchesWon"}).reset_index()
df=pd.merge(team_total_match,team_total_match_won,how='left',left_on=['Venue','City','Team'],right_on=['venue','city','winner'])
percent_win=((df['MatchesWon']/df['MatchesPlayed'])*100).round(2).to_list()
df["PercentWin"]=percent_win
df.fillna(0,inplace = True)
df=df[['Venue','City','Team','MatchesPlayed','PercentWin']].query('PercentWin >= 75 and MatchesPlayed > 5').reset_index()
df=df[['Venue','City','Team','MatchesPlayed','PercentWin']]
df
```

__Brabourne Stadium Mumbai__,__Eden Gardens Kolkata__, __M Chinnaswamy Stadium Banglore__ are the favorite grounds of the __mumbai Indians__ team, they loose only __25%__ of matches on these grounds i.e win percentage is more than equal to __75%.<br>
similarly __Maharashtra Cricket Association Stadium Pune__ for __Chennai Super Kings__ and __Feroz Shah Kotla Delhi__ for __Royal Challengers Bangalore__ are the luckiest ground where they hardly loose any matches

```python
df1=ipl_data[['season','team1','winner']]
df2=ipl_data[['season','team2','winner']]
df2=df2[df2.winner.notnull()]
df1=df1[df1.winner.notnull()]
df2.columns = df1.columns
df=pd.concat([df1, df2],axis=0,sort=False,ignore_index=True)
df.columns=['Season','Team','Winner']
team_total_match=df.groupby("Team")[["Team"]].count().rename(columns={"Team":"Matches Played"}).reset_index()
team_total_match_won=ipl_data.groupby("winner")[["winner"]].count().rename(columns={"winner":"Matches Won"}).reset_index()
df=pd.merge(team_total_match,team_total_match_won,how='inner',left_on='Team',right_on='winner')
percent_win=((df['Matches Won']/df['Matches Played'])*100).round(2).to_list()
df["Percent Win"]=percent_win
df=df.sort_values(['Percent Win'],ascending=False)
df[['Team','Matches Played','Matches Won','Percent Win']]

sns.set(style="whitegrid")
ax = sns.barplot(x='Percent Win', y='Team',data=df)


def show_values_on_bars(axs, h_v="v", space=0.4):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
        
show_values_on_bars(ax, "h", 0.3)        
```

Top three teams wins the table are __Chennai Super Kings__, __Mumbai Indians__ and __Sunrisers Hyderabad__

#### 4.5.4 What are the chances for a final's qualifier to win the finals ?

```python
df1=pd.merge(team_batteling_finals,ipl_data[['date','team1','team2']],how='inner',on='date')[['date','team1','team2']]
df2=df1[['date','team1','team2']]
df2.columns=['date','team2','team1']
df=pd.concat([df1[['date','team1']], df2[['date','team1']]],axis=0,sort=False,ignore_index=True)
final_qualify=df.team1.value_counts()
final_qualify=final_qualify.to_frame()
final_qualify.reset_index(inplace=True)
final_qualify.columns=['Team','QualifyTimes']
df1=pd.merge(team_batteling_finals,ipl_data[['date','season','winner']],how='inner',on=['date','season'])
final_win=df1.winner.value_counts()
final_win=final_win.to_frame()
final_win.reset_index(inplace=True)
final_win.columns=['Team','WinTimes']
df=pd.merge(final_qualify,final_win,how='left',left_on='Team',right_on='Team')
df['Chances']=np.round((df['WinTimes']/df['QualifyTimes'])*100,2)
df.fillna(0,inplace=True)
df.loc[df['QualifyTimes'] >=2].sort_values(['Chances'],ascending=False)
```

Based on the above observation we can say that chances of winning the IPL for __Mumbai Indians__ and __Kolkata Knight Riders__ are quiet high when they qualify for the finals.
